```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
worldPopulation_df=pd.read_csv(r"C:\Users\HIII\OneDrive\Documentos\World Population Growth.csv")
worldPopulation_df.head()
```

```python
worldPopulation_df.info()
worldPopulation_df.describe()
worldPopulation_df.isnull().sum()
```

```python
# Remove commas and convert to 64-bit integers
worldPopulation_df["Population"] = (
    worldPopulation_df["Population"].str.replace(",", "").astype("int64")
)

worldPopulation_df["Number"] = (
    worldPopulation_df["Number"].str.replace(",", "").astype("int64")
)

# Convert Yearly Growth % to float
worldPopulation_df["Yearly Growth %"] = (
    worldPopulation_df["Yearly Growth %"].str.replace("%", "").astype(float)
)

```

```python
worldPopulation_df.dtypes
worldPopulation_df.head()
```

```python
plt.figure(figsize=(10,6))
plt.plot(worldPopulation_df["Year"], worldPopulation_df["Population"], marker="o")
plt.title("World Population Growth Over Time")
plt.xlabel("Year")
plt.ylabel("Population")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(worldPopulation_df["Year"], worldPopulation_df["Yearly Growth %"], color="orange", marker="o")
plt.title("Yearly Population Growth Rate (%)")
plt.xlabel("Year")
plt.ylabel("Growth %")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
sns.lineplot(data=worldPopulation_df, x="Year", y="Density")
plt.title("World Population Density Over Time")
plt.show()

```
